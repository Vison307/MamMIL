```python
import numpy as np

import pandas as pd
import os
```

```python
tasks = ["BRCAS", "BRCAS-VIM", "BRCAS-VMAMBA", 'Camelyon16-transmil', 'Camelyon16-transmil-VIM', 'Camelyon16-transmil-VMAMBA', 'BRCAS-official-VIM', 'BRCAS-official', 'BRCAS-official-VMAMBA']

for task in tasks:
    if task in ['Camelyon16-transmil', 'Camelyon16-transmil-VIM', 'Camelyon16-transmil-VMAMBA', 'BRCAS-official-VIM', 'BRCAS-official', 'BRCAS-official-VMAMBA']:
        log_dir = os.path.join('logs/configs', task)
        rng = range(1)
    else:
        log_dir = os.path.join('logs/configs', task)
        rng = range(10)

    versions = os.listdir(log_dir)
    for version in versions: # each method
        if os.path.isfile(os.path.join(log_dir, version)):
            continue
        seed_rng = [1]
        for seed in seed_rng:
            acc_list, f1_list, recall_list, precision_list, auc_list = [], [], [], [], []
            if len(list(seed_rng)) == 1:
                final_csv_path = os.path.join(log_dir, version, f's{seed}', 'result.csv')
                f = open(final_csv_path, 'w')
                f.write('fold,accuracy,f1score,recall,precision,auc\n')
                for fold in rng:
                    result_csv_path = os.path.join(log_dir, version, f's{seed}', f'fold{fold}', 'result.csv')
                    if not os.path.exists(result_csv_path):
                        continue
                    df = pd.read_csv(result_csv_path)
                    test_Accuracy = df['test_Accuracy'].values[0]
                    test_F1Score = df['test_F1Score'].values[0]
                    test_Recall = df['test_Recall'].values[0]
                    test_Precision = df['test_Precision'].values[0]
                    test_auc = df['test_auc'].values[0]

                    f.write(f'{fold},{test_Accuracy:.4f},{test_F1Score:.4f},{test_Recall:.4f},{test_Precision:.4f},{test_auc:.4f}\n')

                    acc_list.append(test_Accuracy)
                    f1_list.append(test_F1Score)
                    recall_list.append(test_Recall)
                    precision_list.append(test_Precision)
                    auc_list.append(test_auc)

                mean_acc, std_acc = np.mean(acc_list), np.std(acc_list)
                mean_f1, std_f1 = np.mean(f1_list), np.std(f1_list)
                mean_recall, std_recall = np.mean(recall_list), np.std(recall_list)
                mean_precision, std_precision = np.mean(precision_list), np.std(precision_list)
                mean_auc, std_auc = np.mean(auc_list), np.std(auc_list)
                
                if len(rng) == 1:
                    f.write(f'mean,{mean_acc:.4f},{mean_f1:.4f},{mean_recall:.4f},{mean_precision:.4f},{mean_auc:.4f}\n')
                else:
                    f.write(f'mean±std,{mean_acc:.4f}±{std_acc:.4f},{mean_f1:.4f}±{std_f1:.4f},{mean_recall:.4f}±{std_recall:.4f},{mean_precision:.4f}±{std_precision:.4f},{mean_auc:.4f}±{std_auc:.4f}\n')
                f.close()
            else:
                final_csv_path = os.path.join(log_dir, version, 'result.csv')
                f = open(final_csv_path, 'w')
                f.write('fold,accuracy,f1score,recall,precision,auc\n')
                for seed in seed_rng:
                    result_csv_path = os.path.join(log_dir, version, f's{seed}', f'fold0', 'result.csv')
                    if not os.path.exists(result_csv_path):
                        continue
                    df = pd.read_csv(result_csv_path)
                    test_Accuracy = df['test_Accuracy'].values[0]
                    test_F1Score = df['test_F1Score'].values[0]
                    test_Recall = df['test_Recall'].values[0]
                    test_Precision = df['test_Precision'].values[0]
                    test_auc = df['test_auc'].values[0]

                    f.write(f'{seed},{test_Accuracy:.4f},{test_F1Score:.4f},{test_Recall:.4f},{test_Precision:.4f},{test_auc:.4f}\n')

                    acc_list.append(test_Accuracy)
                    f1_list.append(test_F1Score)
                    recall_list.append(test_Recall)
                    precision_list.append(test_Precision)
                    auc_list.append(test_auc)

                mean_acc, std_acc = np.mean(acc_list), np.std(acc_list)
                mean_f1, std_f1 = np.mean(f1_list), np.std(f1_list)
                mean_recall, std_recall = np.mean(recall_list), np.std(recall_list)
                mean_precision, std_precision = np.mean(precision_list), np.std(precision_list)
                mean_auc, std_auc = np.mean(auc_list), np.std(auc_list)
            
                f.write(f'mean±std,{mean_acc:.4f}±{std_acc:.4f},{mean_f1:.4f}±{std_f1:.4f},{mean_recall:.4f}±{std_recall:.4f},{mean_precision:.4f}±{std_precision:.4f},{mean_auc:.4f}±{std_auc:.4f}\n')
                f.close()
                break
            
```
